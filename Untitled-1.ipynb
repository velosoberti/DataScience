```python
from statsmodels.stats.diagnostic import het_white
resid = lm.residuals_
X_com_constante = np.column_stack((np.ones(len(X)), X))
white_statistic, p_valor_white, _, _ = het_white(resid, X_com_constante)
p_valor_white
```

```python
import statsmodels.api as sm

x_constante = sm.add_constant(X)
model = sm.OLS(Y, x_constante).fit()
```

```python
white_statistic, p_valor_white, _, _ = het_white(model.resid, x_constante)
p_valor_white
```

```python
print(model.summary())
```

```python
plt.scatter(X, Y, label='Dados')
plt.plot(X, model.predict(x_constante), color='#8fbcbb', linewidth=2, label='Reta de Regressão')
plt.title('Regressão Linear')
plt.xlabel('Tempo de uso nas Redes Sociais')
plt.ylabel('Mensagens Recebidas')
intercept = model.params[0]
coeficiente = model.params[1]
plt.text(40, 1.65, f'log(y) = {round((intercept),2)} + {round((coeficiente),2)}log(x)', fontsize=12, rotation=31, ha='center'
             , va='center')
```

```python
stat, p_valor = stats.shapiro(model.resid)
plt.figure(figsize=(20,10))
pg.qqplot(model.resid)
plt.annotate(f"Shapiro {p_valor:,.2F}", xy=(0.2,-2))
plt.title('Distribuição dos Resíduos')
plt.show()
```
