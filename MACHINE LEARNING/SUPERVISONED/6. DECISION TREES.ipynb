# Árvore de Decisão

### Algoritmos Versáteis que lidam com Classificações multiclasses e multiouputs e Regressões


#### A árvore de decisão faz decisões a partir de um *root node* [nó da raíz] com a maior pureza (isso significa que a uma das decisões "sim" ou "não" representará apenas um resultado classificatório possível). A partir desse nó a arvore prediz conforme se desce aos nós filhos sendo ***"True"*** as quebras à esquerda e ***"False"*** as quebras à direita 

Obs: Nota-se que as árvores de decisões não exigem tanto pré processamento dos dados como escalonamento ou centralização, etc

- O atributo *sample* dos nós representa a quantidade de itens/instâncias representadas naquele nó

- O atributo *gini* representa a pureza daquele nó, se 0 todas as intâncias do nó se aplicam a classificação

- Você pode optar utilizar o hiperparm ***entropy*** ao invés de ***gini***. Entropia é também uma medida de impureza, a partir dele temos nota da Informação daquele nó. Quanto mais próximo de 0 mais as "mensagens" ou classificações são idênticas. ***Pode-se usar gini ou entropy, na realidade "tanto faz". A melhor interpretação para seus usos segue que Gini costuma isolar mais valores muitos frequentes em seus ramos, já a entropia pode criar árvores mais equilibradas mesmo tendo maior custo computacional***

- Para chamá-lo importamos o *DecisionTreeClassifier* da classe *Tree* do Sklearn

- E setamos o ***max_depth** para nos dar uma quantidade máxima de ramos

- E para melhor visualização da árvore, o sklearn oferece a function *plot_tree* da classe Tree


#### Podemos, também, calcular a probabilidade de cada instância pertencer a determinada classe

Basta usar a função predict_proba() do DecisionTreeClassifier


### Regressão com Árvores de Decisão 

- Para isso usamos a DecisionTreeRegressor

- A única diferença, aqui, é que ao invés de predizer alguma classe ela irá predizer um valor para cada nó

- Aqui a divisão dos nós e ramos se dá pela necessidade de diminuir o MSE

- Regressões tendem a serem bem sobreajustadas, utilize o param *min_samples_leaf* para dar penâltis ao modelo e remover um pouco do sobreajuste

### Hiperparâmetros Da Árvore de Decisão

Quando se deixa a árvore operar de forma livre, é tendencioso a gerar sobreajustes, para isso faz-se necessário alterar os hiperparâmetros do modelo 

Há diversos, como min_samples_split ou min_samples_leaf ou max_samples_split e max_samples_leaf. A ideia é ***aumentar os param min_ ou reduzir os param max_*** para regularizar seu modelo

class sklearn.tree.DecisionTreeClassifier(*, criterion='gini', splitter='best', max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features=None, random_state=None, max_leaf_nodes=None, min_impurity_decrease=0.0, class_weight=None, ccp_alpha=0.0, monotonic_cst=None)

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

iris = load_iris()

x = iris.data[:, 2:]
y = iris.target

tree_clf = DecisionTreeClassifier(max_depth = 2)
tree_clf.fit(x, y)
```

```python
from sklearn.tree import plot_tree

plot_tree(tree_clf, filled=True)
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(max_depth=2)

tree_reg.fit(x, y)
```

```python
plot_tree(tree_reg, filled=True)
plt.show()
```

```python
plt.figure()
clf = DecisionTreeClassifier().fit(iris.data, iris.target)
plot_tree(clf, filled=True)
plt.title("Decision tree trained on all the iris features")
plt.show()
```
